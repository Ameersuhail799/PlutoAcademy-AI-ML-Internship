```python
import pandas as pd
import zipfile

# Unzip the uploaded file
with zipfile.ZipFile('archive.zip', 'r') as zip_ref:
    zip_ref.extractall('netflix_data')

# Read the file
df = pd.read_csv('netflix_data/netflix_titles.csv')

# Show the rows and columns
print(df.shape)
```

```python
df.info()
```

This dataset contains information about all the movies and TV shows available on Netflix, featuring 8,807 entries and 12 columns. The columns describe content traits such as titles, genres, release years, directors, cast members, and production countries. A quick look at the details shows that there are significant missing records in the columns for directors, cast members, and countries. The minor fields tracking ratings and dates added are mostly complete. This structured base will be cleaned next to perform exploratory data analysis.

```python
# Create a safe copy of the data to clean
df_clean = df.copy()

# 1. Fill missing text columns with 'Unknown'
df_clean['director'] = df_clean['director'].fillna('Unknown')
df_clean['cast'] = df_clean['cast'].fillna('Unknown')
df_clean['country'] = df_clean['country'].fillna('Unknown')

# 2. Drop the tiny few rows with missing date or rating
df_clean = df_clean.dropna(subset=['date_added', 'rating'])

# 3. Check if all missing values are fixed (should show 0)
print(df_clean.isnull().sum())
```

```python
# Drop the 3 rows with missing duration values
df_clean = df_clean.dropna(subset=['duration'])

# Check one final time to ensure EVERYTHING is 0
print(df_clean.isnull().sum())
```

### Data Cleaning Documentation
To clean this dataset, all massive text columns missing significant data ('director', 'cast', and 'country') were handled by filling the empty values with the label 'Unknown'. This prevents the loss of valuable rows for overall trends. For the minor columns ('date_added', 'rating', and 'duration') where only a tiny handful of values were missing, the incomplete rows were safely dropped from the dataset. The cleaned dataset now contains zero missing values across all 12 columns, providing a stable foundation for exploratory data analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Re-initialize data path quickly for cell stability
df_clean = pd.read_csv('netflix_data/netflix_titles.csv')
df_clean['country'] = df_clean['country'].fillna('Unknown')
top_countries = df_clean['country'].value_counts().head(5)

# 2. Generate the Bar Chart
plt.figure(figsize=(8, 5))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='viridis')

# 3. Add Mandatory Titles and Labels
plt.title('Top 5 Content Producing Countries on Netflix', fontsize=14, fontweight='bold')
plt.xlabel('Country', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Quickly prepare the timeline data
df_clean = pd.read_csv('netflix_data/netflix_titles.csv')
df_clean['year_added'] = pd.to_datetime(df_clean['date_added'].str.strip()).dt.year
yearly_counts = df_clean['year_added'].value_counts().sort_index()

# 2. Generate the Line Chart
plt.figure(figsize=(9, 5))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o', color='red', linewidth=2)

# 3. Add Titles and Labels
plt.title('Total Content Added to Netflix Over the Years', fontsize=14, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Titles Added', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Quickly isolate the Movie data
df_clean = pd.read_csv('netflix_data/netflix_titles.csv')
movies_only = df_clean[df_clean['type'] == 'Movie']

# 2. Generate the Histogram
plt.figure(figsize=(9, 5))
sns.histplot(data=movies_only, x='release_year', bins=30, kde=True, color='purple')

# 3. Add Mandatory Titles and Labels
plt.title('Distribution of Movie Release Years on Netflix', fontsize=14, fontweight='bold')
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Count of Movies', fontsize=12)
plt.xlim(1940, 2021) # Zooming in on the core movie eras

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Quickly prepare the numeric columns
df_clean = pd.read_csv('netflix_data/netflix_titles.csv')
df_clean['year_added'] = pd.to_datetime(df_clean['date_added'].str.strip()).dt.year
# Drop rows where year_added couldn't be calculated
df_clean = df_clean.dropna(subset=['year_added'])

# 2. Generate the Scatter Plot
plt.figure(figsize=(9, 5))
sns.scatterplot(data=df_clean, x='release_year', y='year_added', alpha=0.5, color='orange')

# 3. Add Mandatory Titles and Labels
plt.title('Movie/Show Release Year vs. Year Added on Netflix', fontsize=14, fontweight='bold')
plt.xlabel('Original Release Year', fontsize=12)
plt.ylabel('Year Added to Platform', fontsize=12)
plt.xlim(1940, 2021)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Quickly isolate the Type counts
df_clean = pd.read_csv('netflix_data/netflix_titles.csv')
type_counts = df_clean['type'].value_counts()

# 2. Generate the Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(type_counts.values, labels=type_counts.index, autopct='%1.1f%%',
        colors=['#E50914', '#221F1F'], startangle=140,
        textprops={'color': 'white', 'weight': 'bold', 'size': 12})

# 3. Add Mandatory Title
plt.title('Percentage Split of Movies vs. TV Shows on Netflix', fontsize=14, fontweight='bold', color='black')

# Set background to transparent or clear so labels show cleanly
plt.gca().set_facecolor('none')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Quickly prepare cross-tab data for top ratings
df_clean = pd.read_csv('netflix_data/netflix_titles.csv')
top_ratings = df_clean['rating'].value_counts().head(5).index
filtered_df = df_clean[df_clean['rating'].isin(top_ratings)]

# Create cross-tabulation table
pivot_table = pd.crosstab(filtered_df['rating'], filtered_df['type'])

# 2. Generate the Heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(pivot_table, annot=True, fmt='d', cmap='Reds', cbar=True)

# 3. Add Mandatory Titles and Labels
plt.title('Heatmap of Content Distribution by Type and Rating', fontsize=14, fontweight='bold')
plt.xlabel('Content Type', fontsize=12)
plt.ylabel('Content Rating', fontsize=12)

# Display the plot
plt.show()
```

## Final Insights Report

1. According to the Pie Chart (Chart 5), Movies overwhelmingly dominate Netflix’s library, accounting for 69.6% of the content compared to just 30.4% for TV Shows.
2. The Bar Chart (Chart 1) shows that the United States is the primary driver of content on the platform with 2,818 titles, followed by India with 972 titles.
3. Based on the Line Chart (Chart 2), content acquisition peaked aggressively in 2019 with over 2,000 titles added, followed by a slight decline during the subsequent years.
4. The Histogram (Chart 3) highlights that Netflix's catalog is strongly skewed toward highly modern releases, with the vast majority of its movies produced after 2015.
5. The Heatmap (Chart 6) confirms that mature audiences are the primary target market, as mature content ('TV-MA') represents the single largest block for both movies and series.

### What Surprised Me the Most
What surprised me the most during this analysis was the sharp difference in content ratings between formats revealed in the heatmap. While mature content ('TV-MA') is massive everywhere, strict movie ratings like 'R' and 'PG-13' have almost zero presence in the TV Show format, showing how differently Netflix categorizes its episodic series compared to traditional films.
